Exercise 1: Understanding Data Visualization

Task: Explain why data visualization is important in data analysis.

Data visualization helps us see the story behind the numbers. It’s way easier to understand trends or comparisons with a chart than with a big table of data. Basically, it makes information clearer, faster to read, and easier to explain to others.

Task: Describe the purpose of a line graph in data visualization.

A line graph is used to show how something changes over time. For example, if you want to see how temperature or prices go up or down, it’s super useful. The line helps you follow the progress like a little visual story.



```python
#Exercise 2: Creating a Line Plot for Temperature Variation

import matplotlib.pyplot as plt
```

```python
Day = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
Temperature = [72, 74, 76, 80, 82, 78, 75]

plt.plot(Day, Temperature, color='pink', linestyle='-', marker='o')
plt.title('Temperature per day')
plt.show()
```

```python
#Exercise 3: Visualizing Monthly Sales with a Bar Chart

month = ["January", "February", "March", "April", "May"]
sales = [5000, 5500, 6200, 7000, 7500]
plt.bar(month, sales, color=['cyan', 'blue', 'darkblue', 'grey', 'black'])
plt.title("Bar Chart: Product Sales")
plt.xlabel("Months")
plt.ylabel("Sales Amount ($)")
plt.show()
```

```python
#Exercise 4: Visualizing the Distribution of CGPA

import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("student_clustering.csv")
```

```python
df = df.drop(columns=['iq'])
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='cgpa', bins=10, color='red', kde=False)


plt.title("Distribution of Students CGPA")
plt.xlabel("CGPA")
plt.ylabel("Amount of Students")
plt.show()
```

```python
#Exercise 5: Comparing Anxiety Levels Across Different Genders

df = pd.read_csv("Student Mental health.csv")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Choose your gender', hue='Do you have Anxiety?', palette='dark:red')


plt.title('Anxiety Levels by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.legend(title='Anxiety')
plt.tight_layout()

plt.show()
```

```python
#Exercise 6: Exploring the Relationship Between Age and Panic Attacks

df = pd.read_csv("Student Mental health.csv")
```

```python
df['Panic_Attack_Number'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})
```

```python
plt.figure(figsize=(8, 5))

plt.figure(figsize=(8, 5))
sns.histplot( data=df, bins=10, x='Age', hue='Panic_Attack_Numeric', multiple='stack', palette='dark:red')

plt.title('Panic Attacks by Age')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.grid(False)

plt.show()
```
