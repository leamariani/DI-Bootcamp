```python
#1. Line Plot Exercise:
#Create a line plot for the function y = x^2 for x ranging from 0 to 10. Add appropriate labels and a title.


# Gnrer des valeurs pour x entre 0 et 10 (100 points)
x = np.linspace(0, 10, 100)

# Calculer y = x
y = x ** 2
```

```python
# Crer le graphique
plt.figure(figsize=(8, 5))            # Taille du graphique
plt.plot(x, y, label='y = x', color='blue')  # Courbe bleue avec lgende
```

```python
# Ajouter un titre et des tiquettes
plt.title('Line Plot of y = x')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Ajouter une grille et une lgende
plt.grid(True)
plt.legend()
```

```python
# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
x = range(0, 11)
y = [i**2 for i in x]
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.title("Line Plot: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], shade=True, color='purple')
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter', color='green')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
sns.barplot(x='class', y='survived', data=titanic, palette='Blues')
plt.title('Survival Rate by Class')
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='Set3')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.violinplot(x='class', y='age', data=titanic, palette='muted')
plt.title('Age Distribution by Class')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=data, aspect=1.5, height=5)
plt.title('Scatter Plot with Regression Line')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg', color='orange')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', palette='Dark2')
plt.show()
```

```python
correlation_matrix = iris.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
data = sns.load_dataset('tips')
g = sns.FacetGrid(data, col='time', row='smoker', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip', color='red')
plt.show()
```

```python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap='Blues')
g.map_diag(sns.histplot, kde=True, color='teal')
g.add_legend()
plt.show()
```

```python
sns.set_context('talk')
sns.set_style('darkgrid')
sns.histplot(data['total_bill'], color='darkblue', bins=15)
plt.title('Advanced Aesthetic Customization')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Exemple avec le dataset iris
iris = sns.load_dataset("iris")

# Trac avec couleur (hue) + forme de marqueur (style) selon l'espce
sns.scatterplot(
    data=iris,
    x='sepal_length',
    y='sepal_width',
    hue='species',           # couleur selon l'espce
    style='species',         # forme de marqueur selon l'espce
    palette='Dark2',         # palette de couleurs
    s=100                    # taille des points
)

plt.title("Iris Dataset - Diffrents marqueurs et couleurs par espce")
plt.grid(True)
plt.show()
```

```python
iris = sns.load_dataset("iris")

correlation = iris.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Heatmap: Correlation Matrix")
plt.show()
```
