```python
#Dans le jeu de données Iris, la plupart des colonnes comme SepalLengthCm, SepalWidthCm, PetalLengthCm et PetalWidthCm sont quantitatives car elles représentent de vraies mesures — on peut les additionner, faire des moyennes, ou comparer les valeurs. La colonne Id est aussi numérique, mais elle sert juste d’étiquette ou de numéro de ligne, donc elle n’a pas vraiment de valeur analytique. En revanche, Species est qualitative car elle indique de quelle espèce de fleur il s’agit (comme Iris-setosa ou Iris-virginica), pas une mesure. On ne peut pas faire de calculs avec les espèces, mais on peut les trier ou les regrouper. En résumé, si c’est un nombre avec lequel on peut faire des maths, c’est quantitatif. Si c’est un mot ou une catégorie, c’est qualitatif.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exercice 5

# Charger depuis seaborn
iris_df = sns.load_dataset("iris").rename(columns={
    "sepal_length": "SepalLengthCm",
    "sepal_width": "SepalWidthCm",
    "petal_length": "PetalLengthCm",
    "petal_width": "PetalWidthCm"
})

# Analyse
sepal_lengths = iris_df['SepalLengthCm']
mean_val = sepal_lengths.mean()
median_val = sepal_lengths.median()
mode_val = sepal_lengths.mode().iloc[0]

# Histogramme
plt.figure(figsize=(8, 5))
sns.histplot(sepal_lengths, kde=True, bins=20, color='skyblue')
plt.title('Distribution of Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

mean_val, median_val, mode_val
```

The distribution of sepal length shows most flowers fall between 5.0–6.5 cm, with fewer very small (<4.8 cm) or very large (>7.0 cm) cases. This suggests the dataset is concentrated around average sizes, making it reliable for classification tasks. Extreme values are rare and can be monitored as potential outliers.
