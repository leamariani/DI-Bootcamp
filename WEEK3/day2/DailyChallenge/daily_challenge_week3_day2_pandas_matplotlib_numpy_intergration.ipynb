```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("global_power_plant_database.csv")
```

```python
info = df.head()
```

```python
info = df.info()
```

```python
info = df.describe()
```

```python
missing_values = df.isnull().sum()
```

```python
# 1. Summary statistics (mean, median, std) for numeric columns
summary_stats = df[
    ['capacity_mw', 
     'generation_gwh_2013', 'generation_gwh_2014', 
     'generation_gwh_2015', 'generation_gwh_2016', 
     'generation_gwh_2017', 'generation_gwh_2018', 
     'generation_gwh_2019']
].agg(['mean', 'median', 'std'])

# Print the summary
print("Summary Statistics:")
print(summary_stats)

# 2. Top 10 countries by number of power plants
plants_by_country = df['country_long'].value_counts().head(10)
print("\nTop 10 Countries by Number of Power Plants:")
print(plants_by_country)

# 3. Top 10 fuel types by number of power plants
plants_by_fuel = df['primary_fuel'].value_counts().head(10)
print("\nTop 10 Fuel Types by Number of Power Plants:")
print(plants_by_fuel)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=plants_by_country.values, y=plants_by_country.index, palette="Blues_d")
plt.title("Top 10 Countries by Number of Power Plants")
plt.xlabel("Number of Power Plants")
plt.ylabel("Country")
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
sns.barplot(x=plants_by_fuel.values, y=plants_by_fuel.index, palette="Greens_d")
plt.title("Top 10 Fuel Types by Number of Power Plants")
plt.xlabel("Number of Power Plants")
plt.ylabel("Fuel Type")
plt.tight_layout()
plt.show()
```
