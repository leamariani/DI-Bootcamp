Exercise 2 : Statistical Analysis
Instructions
In this exercise, youâ€™ll calculate statistical measures for a dataset:

Mean: The average value of a dataset.
Median: The middle value in a dataset when it is arranged in ascending or descending order.
Standard Deviation: A measure of the amount of variation or dispersion in a set of values.
Using NumPy, generate an array of 50 random numbers and compute:

The mean and median.
The standard deviation.

```python
import numpy as np
```

```python
# Generate an array of 50 random numbers (e.g., from a normal distribution)
data = np.random.randn(50)
```

```python
# Compute the mean
mean_value = np.mean(data)
```

```python
# Compute the median
median_value = np.median(data)
```

```python
# Compute the standard deviation
std_deviation = np.std(data)
```

```python
# Print the results
print("Generated Data:", data)
print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot histogram
plt.hist(data, bins=10, edgecolor='black', alpha=0.7)
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='dotted', linewidth=1.5, label=f'Median: {median_value:.2f}')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Boxplot
plt.boxplot(data, vert=False)
plt.title('Boxplot of Random Data')
plt.xlabel('Value')
plt.grid(True)
plt.show()
```

```python
# Line chart
plt.plot(data, marker='o', linestyle='-')
plt.title('Line Chart of Random Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.axhline(np.mean(data), color='red', linestyle='dashed', label='Mean')
plt.axhline(np.median(data), color='green', linestyle='dotted', label='Median')
plt.legend()
plt.grid(True)
plt.show()
```

Exercise 4 : Data Manipulation with NumPy and Pandas
Instructions
Create a DataFrame with random numbers and perform:

Conditional selection of data.
Aggregation functions like sum and average.

```python
# Create random DataFrame
np.random.seed(0)
data = np.random.randint(1, 100, size=(5, 4))
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D'])
```

```python
print("Original DataFrame:")
print(df)
```

```python
condition = df[df['A'] > 50]
print("\nRows where column A > 50:")
print(condition)
```

```python
sum_values = df.sum()
print("\nSum of each column:")
print(sum_values)
```

```python
mean_values = df.mean()
print("\nAverage (mean) of each column:")
print(mean_values)
```

```python
df.plot(kind='bar', figsize=(10, 5))
plt.title('Original Random DataFrame')
plt.xlabel('Row Index')
plt.ylabel('Value')
plt.show()
```

```python
sum_values.plot(kind='bar', color='skyblue', figsize=(6, 4))
plt.title('Sum of Each Column')
plt.ylabel('Sum')
plt.tight_layout()
plt.show()
```

```python
mean_values.plot(kind='bar', color='salmon', figsize=(6, 4))
plt.title('Mean of Each Column')
plt.ylabel('Mean')
plt.grid(True)
plt.tight_layout()
plt.show()
```
