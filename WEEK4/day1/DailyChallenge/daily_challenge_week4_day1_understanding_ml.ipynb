L’apprentissage automatique permet aux ordinateurs d’apprendre à partir de données pour faire des prédictions ou prendre des décisions sans être explicitement programmés pour chaque tâche. 

On imagine apprendre à un enfant à reconnaître des fruits : on montre des pommes, des bananes, etc. Petit à petit, il apprend à les identifier. C’est exactement ce que fait le Machine Learning avec les données. 

Pourquoi c’est utile pour les data analysts ?
Il aide à découvrir des tendances cachées
Il permet de prédire des comportements ou des résultats
Il automatise des tâches répétitives comme la détection d'erreurs ou la classification

Applications de l'apprentissage automatique :
Santé : Prédire les maladies à partir d’analyses médicales.
Banque : Détecter des fraudes sur les cartes bancaires.
Streaming : Recommander des films et séries sur Netflix selon vos goûts.


Les types d'apprentissage automatique

Apprentissage Supervisé : Le modèle apprend à partir de données étiquetées (on connaît déjà la réponse).
-Prédire si un email est spam ou non-spam.
Apprentissage Non-Supervisé : Le modèle apprend sans connaître les réponses. Il trouve des groupes ou structures cachées.
-Segmenter des clients selon leurs comportements.
Apprentissage par Renforcement : Le modèle apprend en recevant des récompenses ou punitions pour ses actions.
-Un robot apprend à éviter des obstacles.


Le processus de développement d’un modèle

1/Sélection des fonctionnalités
On choisit les bonnes colonnes/variables qui influencent le résultat.
Exemple : Pour estimer le prix d’une maison, on garde la surface, le quartier, l’âge de la maison...

2/Sélection du modèle
On choisit un algorithme selon le problème :
-Régression : prédire un nombre
-Classification : prédire une catégorie
-Clustering : regrouper des points similaires

3/Évaluation du modèle
On teste le modèle avec de nouvelles données :
-Précision
-Erreurs (faux positifs, faux négatifs...)
-Performances (F1-score, courbes d’apprentissage)


```python
import matplotlib.pyplot as plt

# Étapes du processus
etapes = [
    "Comprendre le problème",
    "Collecte des données",
    "Nettoyage des données",
    "Sélection des fonctionnalités",
    "Choix du modèle",
    "Entraînement",
    "Évaluation",
    "Optimisation"
]

# Création du graphique
fig, ax = plt.subplots(figsize=(8, 5))
ax.axis('off')

# Afficher chaque étape sous forme de texte avec flèches
for i, etape in enumerate(etapes):
    ax.text(i + 0.5, 0.5, etape, ha='center', va='center',
            bbox=dict(boxstyle="round", facecolor="lightcyan", edgecolor="black"))
    if i < len(etapes) - 1:
        ax.annotate("→", xy=(i + 0.9, 0.5), xytext=(i + 0.7, 0.5),
                    ha='center', va='center', fontsize=30)

plt.title("Étapes principales d'un projet de Machine Learning", fontsize=14)
plt.show()
```
