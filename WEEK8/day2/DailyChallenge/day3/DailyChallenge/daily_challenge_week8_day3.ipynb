# Agent IA pour la Répartition Médicale d’Urgence

## Objectif

Concevoir un agent intelligent capable de :

- Recevoir et analyser des appels d'urgence (voix ou texte)
- Évaluer la gravité des symptômes
- Mémoriser les informations critiques
- Déclencher l’envoi des secours médicaux appropriés

## 1. Définir l’Environnement de l’Agent

### Données perçues (entrées) :

- Transcription de l’appel (voix ou texte)
- Localisation du patient (GPS ou adresse)
- Identité du patient (si connue)
- Antécédents médicaux (si accessibles)

## 2. Outils & API nécessaires

### 1. API de vérification des symptômes

- Entrée : Liste des symptômes
- Sortie : Diagnostic, recommandations, gravité estimée

### 2. Système de gestion des ambulances

- Entrée : Adresse, niveau d’urgence
- Sortie : Ambulance envoyée, ETA

### 3. Modèle de triage médical IA

- Entrée : Texte des symptômes
- Sortie : Score de gravité

## 3. Gestion de l’État

### Schéma JSON :

```json
{
  "appelant": {
    "nom": "Jean Dupont",
    "numéro": "+33612345678",
    "localisation": "12 rue de la Santé, Paris"
  },
  "symptômes": [
    {"symptôme": "douleur thoracique", "gravité": "élevée"}
  ],
  "score_urgence": 92,
  "niveau_urgence": "Élevé",
  "actions": [
    "analyse symptômes",
    "demande ambulance",
    "envoyé confirmation SMS"
  ]
}
```

## 4. Processus de Prise de Décision

### Exemple de logique :

```python
if score_urgence >= 80:
    niveau = "Élevé"
    action = "Envoyer ambulance immédiatement"
elif score_urgence >= 50:
    niveau = "Moyen"
    action = "Conseiller centre de soins"
else:
    niveau = "Faible"
    action = "Donner instructions d’auto-soins"
```

## 5. Type d’Architecture Choisie

### Architecture : Hybride

- Combine réactif + délibératif
- Mémoire de l’état
- Planifie via score et outils

## 6. Comparaison avec un Agent Réactif

| Aspect | Hybride | Réactif |
|--------|---------|---------|
| Mémoire | Oui | Non |
| Planification | Oui | Non |
| Outils | APIs & modèles combinés | Invocation directe |

### Trade-offs :

- Vitesse : Réactif plus rapide
- Fiabilité : Hybride plus robuste
- Intelligence : Hybride plus nuancé

## 7. Réflexion Critique

### Que se passe-t-il sans mémoire ?

L’agent oublierait l’appelant, les actions, les symptômes → système peu fiable.

### Pourquoi les outils sont-ils essentiels ?

Les APIs permettent à l’agent de prendre des décisions médicales fiables et d’agir efficacement avec les secours.
